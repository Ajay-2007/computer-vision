# Grayscaling

#### Grayscaling is process by which an image is conerted from a full color to shades of gray (black & white)
This is done because it simplifies the image, acting almost as a noise reduction and increasing
processing time as there is less information in the image.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def image_show(input_image, format = cv2.COLOR_BGR2RGB):
    plt.imshow(cv2.cvtColor(input_image, format))
    plt.show()
```

```python
# load our input image
input = cv2.imread('./images/input.jpg')
# plt.imshow(cv2.cvtColor(input, cv2.COLOR_BGR2RGB))
# plt.show()
image_show(input, cv2.COLOR_BGR2RGB)
```

```python
# We can use cvtColor, to convert to grayscale
gray_image = cv2.cvtColor(input, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.show()
plt.imshow(input, cmap='gray')
plt.show()
image_show(input, cv2.COLOR_BGR2GRAY)
```
