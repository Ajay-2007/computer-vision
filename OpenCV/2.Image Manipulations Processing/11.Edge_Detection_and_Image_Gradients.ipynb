# Edge Detection & Image Gradients

```
from google.colab import drive
drive.mount('/content/drive')

import cv2
import numpy as np
from IPython.display import display, HTML
import base64
import os

def imshow(name, imageArray):
    # print(name)
    _, png = cv2.imencode('.png', imageArray)
    encoded = base64.b64encode(png)
    return HTML(
        data='''<p>{2}</p><img alt="{0}" src="data:image/png;base64, {1}"/>'''.format(name, encoded.decode('ascii'),
                                                                                      name))

from pathlib import Path
import os
path = Path.cwd() / 'drive' / 'My Drive' / 'Github' / 'computer vision' / 'OpenCV' / 'images'
```

```
image_path = os.path.join(path, 'input.jpg')
# Loading image in grayscale format
image = cv2.imread(image_path, 0)
imshow('Original Image', image )
```

```
height, width = image.shape

# Extract Sobel Edges
sobel_x = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)

imshow('Sobel X', sobel_x)
```

```
imshow('Sobel Y', sobel_y)
```

```
sobel_OR = cv2.bitwise_or(sobel_x, sobel_y)
imshow('sobel_OR', sobel_OR)
```

```
laplacian = cv2.Laplacian(image, cv2.CV_64F)
imshow('Laplacian', laplacian)
```

### Then we need to provide two values
- thresold1 and thresold2. Any gradient value larger than thresold2 is considered to be an edge. Any value below thresold1 is considered not be an edge. 
- Values in between thresold1 and thresold2 are either classified as edges or non-edges based on how their intensities are 'connected'. In this case, any gradient values below 60 are considered non-edges whereas any values above 120 are considered edges.  

```
canny = cv2.Canny(image, 20, 170)
imshow('Canny', canny)
```
