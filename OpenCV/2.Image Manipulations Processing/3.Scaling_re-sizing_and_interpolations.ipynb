## Scaling, re-sizing and interpolations

- Re-sizing is very easy using the cv2.resize function, it's arguments are
`cv2.resize(image, dsize(output image size), x scale, y scale, interpolation)`

```python
import cv2
import numpy as np
from IPython.display import display, HTML
import cv2
import base64
import matplotlib.pyplot as plt
%matplotlib inline 
import os

def imshow(name, imageArray):
    # print(name)
    _, png = cv2.imencode('.png', imageArray)
    encoded = base64.b64encode(png)
    return HTML(
        data='''<p>{2}</p><img alt="{0}" src="data:image/png;base64, {1}"/>'''.format(name, encoded.decode('ascii'),
                                                                                      name))
```

```python
path  = os.path.join('./../images', 'input.jpg')
image = cv2.imread(path)
```

```python
# Let's make out image 3/4 of it's original size
image_scaled = cv2.resize(image, None, fx=0.75, fy=0.75)
imshow('Scaling - Linear Interpolation', image_scaled )
```

```python
# Let's double the size of our image
img_scaled = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
imshow('Scaling - Cubic Interpolation', img_scaled)
```

```python
print(image.shape)
print(img_scaled.shape)

```

```python
# Let's skew the re-sizing by setting exact dimensions
img_scaled = cv2.resize(image, (900, 400), interpolation = cv2.INTER_AREA)
imshow('Scaling - Skewed Size', img_scaled)


```

```python
# image compressed horizontally 
image_scaled = cv2.resize(image, None, fx=0.75, fy=1)
imshow('Scaling - Linear Interpolation Compressed horizontally', image_scaled )

```
