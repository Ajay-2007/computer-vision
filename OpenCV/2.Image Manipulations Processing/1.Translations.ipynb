### Translations

- Affine Transform simply shifts the position of an image.
- We use cv2.warpAffine to implement these transformations.

```python
import cv2
import numpy as np
from IPython.display import display, HTML
import cv2
import base64
import matplotlib.pyplot as plt
%matplotlib inline 
import os

def imshow(name, imageArray):
    # print(name)
    _, png = cv2.imencode('.png', imageArray)
    encoded = base64.b64encode(png)
    return HTML(
        data='''<p>{2}</p><img alt="{0}" src="data:image/png;base64, {1}"/>'''.format(name, encoded.decode('ascii'),
                                                                                      name))
```

```python
path  = os.path.join('./../images', 'input.jpg')
image = cv2.imread(path)
```

```python
# original image
imshow('Original Image', image)
```

```python
# Store height and width of the image
height, width = image.shape[:2]

quarter_height, quearter_width = height / 4, height / 4

#       | 1 0 Tx |
# T =   | 0 1 Ty |
# T is our translation matrix
T = np.float32([[1, 0, quearter_width], [0, 1, quarter_height]])

# We use cv2.warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (width, height)) 
imshow('translated image', img_translation)

```
