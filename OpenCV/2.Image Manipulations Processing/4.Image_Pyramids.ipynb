## Image Pyramids
- Useful when scaling images in object detection.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import base64
from IPython.display import display, HTML

%matplotlib inline


def imshow(name, imageArray):
    # print(name)
    _, png = cv2.imencode('.png', imageArray)
    encoded = base64.b64encode(png)
    return HTML(
        data='''<p>{2}</p><img alt="{0}" src="data:image/png;base64, {1}"/>'''.format(name, encoded.decode('ascii'),
                                                                                      name))

```

```python
path = os.path.join('./../images', 'input.jpg')
image = cv2.imread(path)
```

```python
smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(smaller)
imshow('Original Image', image)
```

```python
imshow('Smaller Image', smaller)
```

```python
imshow('Larger Image', larger)
```

```python
print('Original Image shape',image.shape)
print('Smaller Image shape', smaller.shape)
print('Larger Image shape', larger.shape)
```
