```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input = cv2.imread("./images/input.jpg")

# cv2.imshow("Hello World", input)
plt.imshow(cv2.cvtColor(input, cv2.COLOR_BGR2RGB))
plt.show()
# 'waitkay' allows us to input information when a image window is open
# By leaving it blank it just waits for anykey to be pressed before continuing.
# By placing numbers (except 0), we can specify a delay for
# how long you keep the window open (time is in milliseconds here)
cv2.waitKey()

# This closes all open windows
# Failure to place this will cause your program to hang
cv2.destroyAllWindows()
```

## How images are stored in a Computer

```python
# Shape gives the dimensions of the image array
# The 2D dimensions are 415 pixels in high and 622 pixels wide. The 3 means
# there are 3 other components (or 3 channels) RGB that make up this image.
print(input.shape)

```

```python
print('Height of Image: ', int(input.shape[0]), ' pixels')
print('Width of Image: ', int(input.shape[1]), ' pixels')
```

### How do we save images we edit in OpenCV?

```python
# Simply use 'imwrite' specifying the file name and the image to be saved
cv2.imwrite('./output_images/output.jpg', input)
cv2.imwrite('./output_images/output.png', input)
```
